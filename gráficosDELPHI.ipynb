```
# importando os pacotes necessários
import pandas as pd 
import numpy as np
pd.options.mode.chained_assignment = None  # default='warn'
from bokeh.models import PreText
from bokeh.models import Div
from bokeh.io import curdoc
from bokeh.io import output_file, show
from bokeh.models import Panel, Tabs
from bokeh.plotting import figure
from bokeh.io import show, output_notebook
from bokeh.models import CheckboxGroup
from bokeh.models import Select
from bokeh.models import PrintfTickFormatter
from bokeh.layouts import gridplot
from bokeh.layouts import column
from bokeh.models import PreText
from bokeh.models import ColumnDataSource, DataTable, DateFormatter, TableColumn
from bokeh.models import CheckboxGroup, CrosshairTool, HoverTool
from scipy.integrate import odeint
from datetime import date
from datetime import datetime
import warnings
warnings.filterwarnings('ignore')
```

```
def dados(UF):
  dfB = pd.read_csv("https://raw.githubusercontent.com/maboselli/DELPHI_figs/master/Brasil_" + UF + "_Estado.csv")

  dfB.Day = pd.to_datetime(dfB.Day)

  b1= figure(title="Curva de Casos Acumulados - " + UF,x_axis_type="datetime", plot_width=750, plot_height=400) 

  b1.line(x= dfB.Day, y = dfB['Total Detected'],line_width = 3, color = 'crimson', legend_label = "Curva Estimada")

  #b1.vbar(x=dfCA_CE.Data, width=0.5, bottom=0,top=dfCA_CE.CACiE_An, color="royalblue", legend_label="Casos Acumulados")

  b1.xaxis.axis_label = "Data"
  b1.yaxis.axis_label = "Casos Acumulados"
  b1.yaxis.axis_label_text_font_style = "normal"
  b1.xaxis.axis_label_text_font_style = "normal"
  b1.yaxis.minor_tick_line_color = None
  b1.yaxis[0].formatter = PrintfTickFormatter(format="%f")
  b1.legend.location = "top_left"

  b1.add_tools(HoverTool(
      tooltips=[
          ( 'Casos Acumulados ', '$y{0}' ),
          ( 'Data',   '@x{%F}'            ),# use @{ } for field names with spaces
      ],
      formatters={
          'x'      : 'datetime', 
      },))

  b3 = figure(title = "Óbitos Acumulados - " + UF, x_axis_type = "datetime", plot_width = 750, plot_height = 400)
              
  b3.line(x = dfB.Day, y = dfB['Total Detected Deaths'], line_width = 3, color = "crimson", legend_label = "Curva Estimada")

  #b3.vbar(x=dfOA_CE.Data, width=0.5, bottom=0, top=dfOA_CE.OACiE_An, color="royalblue", legend_label="Óbitos Acumulados")

  b3.xaxis.axis_label = "Data"
  b3.yaxis.axis_label = "Óbitos Acumulados"
  b3.yaxis.axis_label_text_font_style = "normal"
  b3.xaxis.axis_label_text_font_style = "normal"
  b3.yaxis.minor_tick_line_color = None
  b3.yaxis[0].formatter = PrintfTickFormatter(format="%f")
  b3.legend.location = "top_left"


  b3.add_tools(HoverTool(
      tooltips=[
          ( 'Óbitos Acumulados ', '$y{0}' ),
          ( 'Data',   '@x{%F}'            ),# use @{ } for field names with spaces
      ],
      formatters={
          'x'      : 'datetime', 
      },))


  q = Div(text= """<h1> Modelo DELPHI</h1>""",
    width=1000, height=50,)

  q1 = Div(text= """<h2> Curva de Casos Acumulados</h2>""",
  width=1000, height=50,)


  q3 = Div(text= """<h2> Curva de Óbitos Acumulados</h2>""",
  width=1000, height=50,)

  q4 = Div(text= """<h2> Variação do R_t </h2>""",
  width=1000, height=50,)

  q5 = Div(text= """<h2> Tabela com valores simualdos nos próximos 30 dias </h2>""",
  width=1000, height=50,)
  output_file("Delphi" + UF + ".html")
  output_notebook()
  show(column(q, q1,b1,q3,b3))
    
  return

for UF in ["BA", "RN", "MA","AL", "PI","PE","PB", "SE", "CE"]:
  dados(UF)
```
